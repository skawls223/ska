```python
# 

# import the necessary packages
from skimage.exposure import rescale_intensity
import numpy as np
import argparse
import cv2 as cv
from matplotlib import pyplot as plt


def convolve(image, kernel):
    #    3    :2  
    (iH, iW) = image.shape[:2]
    (kH, kW) = kernel.shape[:2]
 
 
    pad = int((kW - 1) / 2)  #  () -> int() 
    image = cv.copyMakeBorder(image, pad, pad, pad, pad, cv.BORDER_REPLICATE)
    output = np.zeros((iH, iW), dtype="float32")
    
    
    for y in np.arange(pad, iH + pad): #np.( )    
        for x in np.arange(pad, iW + pad):
            
           
            roi = image[y - pad:y + pad + 1, x - pad:x + pad + 1] #roi 3x3x3 rgb 

            k = (roi * kernel).sum() #kernel 

            output[y - pad, x - pad] = k

    output = rescale_intensity(output, in_range=(0, 255))
    output = (output * 255).astype("uint8") 
      

    return output

img = cv.imread('s6.jpeg',0)

#kernel define
kernel = np.array([[-1,-1,-1],
                   [-1,8,-1],
                   [-1,-1,-1]])

print(kernel)

dst = convolve(img, kernel)


cv.imshow('img',dst)
cv.imshow('img_ori', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 

# import the necessary packages
from skimage.exposure import rescale_intensity
import numpy as np
import argparse
import cv2 as cv
from matplotlib import pyplot as plt


def convolve(image, kernel, pad, stride):

    (iH, iW) = image.shape[:2]
    (kH, kW) = kernel.shape[:2]
    stri = stride
    padding = int((kW - 1) / 2)
    
    output_width = int((iW+2*pad-kW)/stri+1)
    output_height = int((iH+2*pad-kH)/stri+1)
 
    image = cv.copyMakeBorder(image, pad, pad, pad, pad, 0)
    output = np.zeros((output_height, output_width), dtype="float32")
   
    
    for y in np.arange(padding, int((iH+2*pad-kH)/stri+1)): 
        for x in np.arange(padding, int((iW+2*pad-kW)/stri+1)):
            
           
            roi = image[int((y*stri) - padding):int((y*stri) + padding + 1), int(x*stri - padding):int(x*stri + padding + 1)] 
           
           
            k = (roi * kernel).sum()
            
            output[y-padding, x-padding] = k
   

    output = rescale_intensity(output, in_range=(0, 255))
    output = (output * 255).astype("uint8") 
      
 
    return output

img = cv.imread('s6.jpeg',0)

#kernel define
kernel = np.array([[-1,-1,-1],
                   [-1,8,-1],
                   [-1,-1,-1]])


dst = convolve(img, kernel, 80, 1)


cv.imshow('img',dst)
cv.imshow('img_ori', img)
cv.waitKey(0)
cv.destroyAllWindows()
```
